```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder()
data['histogram_tendency']=oe.fit_transform(data[['histogram_tendency']])
```

```python
data.histogram_tendency.value_counts()
```

```python
data['fetal_health'] = data['fetal_health'].replace({
        1.0: "NORMAL",
        2.0: "SUSPECT",
        3.0: "PATHOLOGICAL"
    })
```

```python
data.head()
```

```python
x=data.drop(['fetal_health'],axis=1)
y=data['fetal_health']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

from sklearn.metrics import classification_report,accuracy_score
```

```python
model={
    'Logisticregression':LogisticRegression(),
    'knn':KNeighborsClassifier(n_neighbors=2),
    'Randomforest':RandomForestClassifier(),
    'Decisiontree':DecisionTreeClassifier(),
    'svc':SVC()
}

for name,model in model.items():
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(name+" : {}%".format(accuracy_score(y_pred,y_test)*100))
    print(name+classification_report(y_pred,y_test))
```

```python
from sklearn.metrics import confusion_matrix
ax= plt.subplot()
sb.heatmap(confusion_matrix(y_test, y_pred), annot=True, ax = ax, cmap = "BuGn");
ax.set_title("Confusion Matrix"); 
ax.xaxis.set_ticklabels(["Normal", "Suspect", "Pathological"]);
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf=RandomForestClassifier()
```

```python
param_grid={
    'n_estimators':[200,500],
    'max_features':['auto','sqrt','log2'],
    'max_depth':[4,5,6,7,8],
    'criterion':['gini','entropy']
}
```

```python
CV_rfc=GridSearchCV(estimator=rf,param_grid=param_grid,cv=5)
CV_rfc.fit(x_train,y_train)
```

```python
CV_rfc.best_params_
```

```python
rfc1=RandomForestClassifier(random_state=42,max_features='sqrt',criterion='entropy',max_depth=8,n_estimators=200)
```

```python
rfc1.fit(x_train,y_train)
```

```python
pred=rfc1.predict(x_test)
```

```python
print('Accuracy for Randomforest:',accuracy_score(y_test,pred)*100)
```
